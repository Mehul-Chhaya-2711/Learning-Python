```python
# Modify the program from the previous exercise to print the 20 most frequently-used
# words in the book.

invert_dictionary = {}
def dict_invert(dictionary):
    """
    Original Dictionary is : Keys are letters in a word and Values are their frequencies in the word
    Inverted Dictionary is: Keys are the frequencies of the letters and values are a list of letters appearing as per
    the frequency of their key indec value.
    
    Steps:
    1. Iterate through dictionary keys
    2. Assign the key's value to a variable
    3. IF that new variable (key's value) is not found as a key in new dictionary then make that new variable as key of new dictionary and
        apply original dictionary's key as value to it. So basically we swapped the roles of jey and value of original dictionary
    4. If that new variable (Value of key in original dictionary) is already present as a key in new dictionary, then append the 
        corresponding key of original dictionary as an element of list which will act as value to the new key
    """
    for keys in dictionary:
        new_value = dictionary[keys]
        if new_value not in invert_dictionary:
            invert_dictionary[new_value] = [keys]
        else:
            invert_dictionary[new_value].append(keys)
    return(invert_dictionary)  

inverted_dick = dict_invert(word_list_unique_count)
sorted_wordlist = sorted (inverted_dick.keys())  #sorted return with a list of sorted indexes which further needs to be used in dictionary
sorted_wordlist.sort(reverse=True)   #reverse sorting the list to get largest to lowest REPITITION index of words

repitition_index_list = []
for i in range(20):
    repitition_index_list.append(sorted_wordlist[i])

    
print('Following are the top 20 most frequent words used in Charles Dickens A tale of 2 cities')    
for i in repitition_index_list:
    print(i,inverted_dick[i])
```

```python
# A FINAL EXERCISE TO END CONTAINERS (LISTS, DICTIONARIES, TUPLES AND SET) AND LEARN WHEN TO USE EACH.

#OBJECTIVE :  BUILD A RANDOM SENTENCE GENERATOR

#A general sentence structure would be something of a combination as in order 
#[Interjection, verb, verb, pronoun, preposition, adjective, noun, 
#adjective, noun, conjunction, adverb, adverb, pronoun, verb, preposition, article, noun, verb.]

# Interjection is always at the end of the sentences
# Verb preceeds pronoun and vice versa
# pronoun always preceeds preposition which preceeds adjective which preceeds noun 
# noun also can preceed adjective as well as a verb.

import random

#STEP 1 : GET A LIST OF RANDOM WORDS FOR ALL 8 PARTS OF SPEECH
#As we here are going to user mapping i.e. get a random word out of many belonging to a PART OF SPEECH : WE WILL USE DICTIONARY
    
PARTS_OF_SPEECH = {
'noun':('bunch','audience','flock','team','group','family','band','village','cat','sock','ship','hero','monkey','baby','match','shahrukh','tom cruise','robert')
,'pronoun':('I','me','he','she','herself','you','it','that','they','each','few','many','who','whoever','whose','someone','everybody')
,'verb':('Go','Grab','Help','Hit','Hop','Shout','Sing','Sit','Smile','kill','enter','fall','cry','call')
,'adjective':('attractive','bald','beautiful','chubby','clean','dazzling','drab','elegant','fancy','fit','flabby','glamorous','gorgeous','handsome','long','magnificent','muscular','plain','plump','quaint','scruffy','shapely','short','skinny','stocky','ugly','unkempt','unsightly')
,'adverb':('not','also','very','often','however','too','usually','really','early','never','always','sometimes','together','likely','simply','generally','instead','actually')
,'preposition':('above','across','against','along','among','around','at','before','behind','below','beneath','beside','between','by','down','from','in','into','near','of','off','on','to','toward','under','upon','with','within')
,'conjunction':('and','that','but','or','as','if','when','than','because','while','where','after','so','though','since','until','whether','before','although','nor','like','once','unless','now','except')
,'interjection':('yes','oh','yeah','no','hey','hi','hello','hmm')
}

#STEP 2: CREATE A FUNCTION TO GET RANDOM WORD OUT OF A LIST

def random_word(list_name):
    return random.choice(list_name)

#STEP 3: CREATE PREFIX SUFFIX RULES FOR PART OF SENTENCES: LIKE IF SENTENCE STARTS FROM NOUN WHAT COULD BE THE SUFFIX THEN

PREFIX_SUFFIX_MAPPING = {
'noun':('conjunction','verb','adjective')
,'pronoun':('verb','adjective')
,'verb':('adjective')
,'adjective':('noun','preposition','adverb')
,'adverb':('verb','adjective')
,'preposition':('noun','pronoun')
,'conjunction':('noun','pronoun')
,'interjection':('noun','pronoun','verb','adjective','adverb','preposition','conjunction')    
}

#STEP 4: CREATE A FUNCTION THAT WRITES SENTENCES WITH DESIRED NUMBER OF WORDS IN IT.

# FUNCTION FLOW
# 1. SELECT A RANDOM PART OF SPEECH >> SELECT THE WORD
# 2. SELECT THE SUFFIX PART OF SPEECH >> SELECT THE WORD
# 3. CONCATENATE THE WORD TO A LARGER STRING
# 4. REPEAT


def sentence(word_count):
    sentence_final = ''
    for i in range(word_count):
        part_of_speech_main = random_word(list(PREFIX_SUFFIX_MAPPING.keys()))
        #print('part_of_speech_main',part_of_speech_main)
        word_one = random_word(PARTS_OF_SPEECH[part_of_speech_main])
        #print('word_one',word_one)
        suffix_part_of_speech = random_word(list(PREFIX_SUFFIX_MAPPING[part_of_speech_main]))
        #print('suffix_part_of_speech',suffix_part_of_speech)
        word_two = random_word(PARTS_OF_SPEECH[part_of_speech_main])
        #print('word_two',word_two)
        sentence_final = sentence_final +' '+ word_one + ' ' + word_two
    return(sentence_final)

print(sentence(6))     


#this program is a very trivial version. Nontheless it can be expanded to create much better sentences
```
