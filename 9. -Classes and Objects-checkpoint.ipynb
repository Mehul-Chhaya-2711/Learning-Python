```python
#THEORY

#CREATING OUR OWN BUILT IN TYPES SUCH AS MATH
#Almost everything in Python is an object, with its properties and methods. 
#A Class is like an object constructor, or a "blueprint" for creating objects.

#built in types
import math
print('print(type(math))',type(math))
print('print(type(math.pi))',type(math.pi))

#created by us
class point(object):
    """Represents a point in 2D space"""
blank = point()
print('point',point,'\n')    
print('type(point)',type(point),'\n')    
print('blank','This is an instance of the class point and not point itself located at indicated memory address',blank,'\n')
print('print(type(blank))','This again is a of class type only',type(blank),'\n')

#as pi is an attribute of math module we will have 2 positional attributes to our class's instance blank

blank.x = 3
blank.y = 4


print('print(type(blank.x)):  an object of',type(blank.x),' With Value: ',blank.x,'\n')
print('print(type(blank.y)):  an object of',type(blank.y),' With Value: ',blank.y,'\n')

#What is blank.x ? It commands the code to go to object blank and fetch the value of attribute x
#This is similar to math.pi ! Go to the module math and fetch the value of float type attribute pi

# Small Exercise: Write a function called distance_between_points that takes two Points as arguments
# and returns the distance between them. The arguments should be our object attributes

def distance_between_points(x,y):
    return(math.sqrt(x**2+y**2))
           
print('distance_between_points is: ',distance_between_points(blank.x,blank.y),'\n')


#Now lets classes pointing to other classes or rather a Class' instance pointing to another's

class Rectangle:
    """This class is a blueprint that will help draw a Rectangle
    
    Attributes in it will be width, height and a corner_point indicator of the rectangle"""
    
#lets create an instance of this class and assign it values

rectangle_inst = Rectangle()
rectangle_inst.width = 100          #refer to rectangle object and assign value to its width attribute
rectangle_inst.height = 200         #refer to rectangle object and assign value to its height attribute
rectangle_inst.corner = point()     #refer to rectangle object and assign value to its center attribute
rectangle_inst.corner.x_coordinate = 3 #again to the center attribute of object rectangle assign to its point object the x coordinate value 
rectangle_inst.corner.y_coordinate = 4 #again to the center attribute of object rectangle assign to its point object the y coordinate value

#point() in object rectangle as rectangle_inst.corner = point()   is called an embedded object. It is being referred to by another object

#LEarning Point: An entire object can be passed as an argument
#lets create a function that finds center point of rectangle and returns the co-ordinates as values
def find_center(rectangle_inst):      #passing an entire object's instance to the function
    point_c = point()
    point_c.x = rectangle_inst.corner.x_coordinate + rectangle_inst.width/2.0 #go half way horizontally from corner point
    point_c.y = rectangle_inst.corner.y_coordinate + rectangle_inst.height/2.0 #go half way vertically from corner point
    return(point_c)

print('Printing the x and y co-ordinates of rectangle',find_center(rectangle_inst).x,find_center(rectangle_inst).y,'\n')

#let us try the same with a built in example by passing entire math module to a function

def print_pi(math):
    pie_value = math.pi
    return(pie_value)

print('Here we pass the math as an entire module to function and then get printed one of its attributes pi',print_pi(math),'\n')



#objects are mutable as in their attribute values can be changed

#A small exercise
# Write a function named move_rectangle that takes a Rectangle and two numbers
# named dx and dy. It should change the location of the rectangle by adding dx to the x coordinate of
# corner and adding dy to the y coordinate of corner. Validate the answer by printing center coordinates of the new rectangles

def move_rectangle(rectangle,dx,dy):
    rectangle.corner.x_coordinate += dx
    rectangle.corner.y_coordinate += dy
    return(find_center(rectangle).x,find_center(rectangle).y)

print('value before change: ',find_center(rectangle_inst).x,find_center(rectangle_inst).y,'\n')

for i in range(10):
    print(move_rectangle(rectangle_inst,50,50),'  ',type(move_rectangle(rectangle_inst,50,50)),' ') 
    #only the thing being returned is of tuple type. The object's internal attributes are still float and mutable
    
print('\n','value after runs: ',find_center(rectangle_inst).x,find_center(rectangle_inst).y,'\n')    




##CREATING COPIES OF INSTANCES; USING SAME FUNCTIONS WITHOUT AFFECTING THE ORIGINAL CLASS INSTANCES
##USE COPY COMMAND COPY.DEEPCOPY

# Write a version of move_rectangle that creates and returns a new Rectangle
# instead of modifying the old one.

import copy
rectangle_new = copy.deepcopy(rectangle_inst)

def move_rectangle_1(rectangle,dx,dy):
    rectangle_new.corner.x_coordinate += dx
    rectangle_new.corner.y_coordinate += dy
    return(find_center(rectangle_new).x,find_center(rectangle_new).y)

print('value before change: ',find_center(rectangle_inst).x,find_center(rectangle_inst).y,'\n')

for i in range(10):
    print(move_rectangle_1(rectangle_inst,50,50),'  ',type(move_rectangle_1(rectangle_inst,50,50)),' ') 
    #only the thing being returned is of tuple type. The object's internal attributes are still float and mutable
    
print('\n','value after runs: ',find_center(rectangle_inst).x,find_center(rectangle_inst).y,'\n')   

#as we can see the values of rectangle_inst remains same as before
```
