```python
#Inverting a dictionary
#Changing key to value and value to key

dictionary = {}

def histogram(a):
    for i in a:
        dictionary[i] = 1 + dictionary.get(i,0)
    return(dictionary) 

histogram('hippopotamus')
print('Following is the dictionary/n')
print(dictionary)

invert_dictionary = {}
def dict_invert(dictionary):
    """
    Original Dictionary is : Keys are letters in a word and Values are their frequencies in the word
    Inverted Dictionary is: Keys are the frequencies of the letters and values are a list of letters appearing as per
    the frequency of their key indec value.
    
    Steps:
    1. Iterate through dictionary keys
    2. Assign the key's value to a variable
    3. IF that new variable (key's value) is not found as a key in new dictionary then make that new variable as key of new dictionary and
        apply original dictionary's key as value to it. So basically we swapped the roles of jey and value of original dictionary
    4. If that new variable (Value of key in original dictionary) is already present as a key in new dictionary, then append the 
        corresponding key of original dictionary as an element of list which will act as value to the new key
    """
    for keys in dictionary:
        new_value = dictionary[keys]
        if new_value not in invert_dictionary:
            invert_dictionary[new_value] = [keys]
        else:
            invert_dictionary[new_value].append(keys)
    print(invert_dictionary)  

print('Following is the inverted dictionary/n')    
dict_invert(dictionary)
```

```python
# FACT

# Lists can be values in dictionaries but not keys. Keys can only be a single value such as integer, float etc but not list.

# Here's why!!

# Keys are hashable. Meaning they are provided with a specific integer to whom the program would look up to get corresponding values.
# This system works fine if the keys are immutable. But if the keys are mutable, like lists,
# bad things happen. For example, when you create a key-value pair, Python hashes the key
# and stores it in the corresponding location. If you modify the key and then hash it again, it
# would go to a different location. In that case you might have two entries for the same key,
# or you might not be able to find a key. Either way, the dictionary wouldn’t work correctly.
# That’s why the keys have to be hashable, and why mutable types like lists aren’t. The
# simplest way to get around this limitation is to use tuples, which we will see in the next
# chapter.
```

```python
#MEMOS

#A CLASSIC EXMAPLE OF WHERE CAN WE USE DICTIONARIES.

def fibonacci (n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print(fibonacci(0))   

#HOW TO ADD VALUES TO DICTIONARY? SIMPLY BY DICT[KEY] = VALUE WILL ADD A KEY:VALUE PAIR TO IT

#LET US MAKE A KEY:VALUE PAIR FOR FIBONNACI RESULTS OF 0 AND 1

known = {0:0,1:1}

def fibonacci1(n):
    """Steps
    1. IF we know the fibonnaci value of n then search for its value in dictionary 'known' and immediately return the answer
    instead of running the entire funcion
    2. This way when we have a dictionary of many low level fibonnaci numbers and their values we do not need to recurse 
    through function and can save a lot of processing time by simple returning the known value rather than rerunning the function"""
    if n in known:
        return known[n]
    res = fibonacci(n-1) + fibonacci(n-2)
    known[n] = res
    return res

print(fibonacci1(50))  
print(known)
```

```python
# In a given list What is the longest English word, that remains a valid English word, as you remove its
# letters one at a time?

# Now, letters can be removed from either end, or the middle, but you can’t rearrange any
# of the letters. Every time you drop a letter, you wind up with another English word. If
# you do that, you’re eventually going to wind up with one letter and that too is going
# to be an English word—one that’s found in the dictionary. I want to know what’s the
# longest word and how many letters does it have?
# I’m going to give you a little modest example: Sprite. Ok? You start off with sprite,
# you take a letter off, one from the interior of the word, take the r away, and we’re left
# with the word spite, then we take the e off the end, we’re left with spit, we take the s off,
# we’re left with pit, it, and I.

# HINT

# 1. You might want to write a function that takes a word and computes a list of all the words that
# can be formed by removing one letter. These are the “children” of the word.
# 2. Recursively, a word is reducible if any of its children are reducible. As a base case, you can
# consider the empty string reducible.
# 3. The wordlist I provided, words.txt, doesn’t contain single letter words. So you might want
# to add “I”, “a”, and the empty string.
# 4. To improve the performance of your program, you might want to memoize the words that are
# known to be reducible.

y = ('sprite','rite','spite','site','spit','pit','i','sit')

word_to_children_dict = {}

def pop_letter(string,indexx):
    word_list = list(string)
    word_list.pop(indexx)
    return_word = ''
    for i in word_list:
        return_word = return_word + i
    return return_word

def cut_letter_form_word(word):
    #including global variables in function so that changes made under the function are alive else where in the program too
    global word_to_children_dict
    for i in range(len(word)):
        children_words = []
        if pop_letter(word,i) in y:
            children_words.append(pop_letter(word,i))
            cut_letter_form_word(pop_letter(word,i))        
                       
def longest_english_word(word):
    cut_letter_form_word(word)
    word_to_children_dict[word] = children_words
    
    
longest_english_word('sprite')
print(children_words)   
print(word_to_children_dict)  
print(len(word_to_children_dict['sprite']))
#CONTINUED BELOW FOR COMPLETE SOLUTION OF FINDING THE LONGEST WORDS
```
